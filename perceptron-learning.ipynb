```python
# Line Drawing
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 10])
ypoints = np.array([0, 20])

plt.plot(xpoints, ypoints, 'r')
plt.grid()
plt.show() 
```

```python
Scatter Plot
```

```python
x = np.array([1,1,3,1,9,4,2,6,6])
y = np.array([1,2,4,9,2,3,1,5,8])
plt.scatter(x, y, color = 'blue')

plt.grid()
plt.show()
```

```python
xpoints = np.array([0, 10])
ypoints = np.array([0, 25])

plt.plot(xpoints, ypoints, 'r')

x = np.array([1,1,3,1,9,4,2,6,6])
y = np.array([1,2,4,9,2,3,1,5,8])
plt.scatter(x, y, color = 'green')

plt.grid()

plt.show()
```
